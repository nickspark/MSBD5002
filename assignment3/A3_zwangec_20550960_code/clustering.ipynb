### read feature from feature_create.py

```python
import numpy as np
import sys
import pandas as pd

features = pd.read_pickle('resnet_v1_101_features.pkl')
features
    
```

### create feature dataframe

```python
df = pd.DataFrame(features)
df = df.T
df = df.sort_index()

df
```

### draw hierarchy cluster picture to observe the number of clusters

```python
import scipy
import scipy.cluster.hierarchy as sch
import matplotlib.pylab as plt
import matplotlib
disMat = sch.distance.pdist(df,'euclidean') 
Z=sch.linkage(disMat,method='average') 
```

```python
fig = plt.figure(figsize=(25, 10))
dn = sch.dendrogram(Z)
plt.show()
fig.savefig('plot_dendrogram.png', dpi = 100)
```

```python
cluster= sch.fcluster(Z, t=60, criterion = 'distance') 

num_clusters = cluster.max()
indices = []
for cluster_number in range(1, num_clusters + 1):
    indices.append(np.where(cluster == cluster_number)[0])
dic = np.zeros(5011)
for num in range(len(indices)):
    for i in indices[num]:
        dic[i] = int(num)
        
result = indices
result
```

```python
num = len(result)
num
```

### the hierarchy clustering's result is not good, for so many cluster only have one picture, so after choose a k, we use kmeans 

```python
from sklearn.cluster import KMeans
train_X=df.values
kmeans = KMeans(n_clusters=23, random_state=0).fit(train_X)

res = list(kmeans.predict(train_X))
```

```python
clus = []
for i in range(num):
    temp = []
    for a in range(len(res)):
        if res[a] == i:
            temp.append(str(a).zfill(5))
    clus.append(temp)
for i in clus:
    print(i)
```
